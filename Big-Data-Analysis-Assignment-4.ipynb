```python
filename = 'sample.fastq'
f = open(filename,'r')
fastq = f.readlines()
```

**Count the number of each base across all sequences**

```python
counter={'A':0,'C':0,'G':0,'T':0}
for line_number,line in enumerate(fastq):
    #if it's a sequencing read: if it's line 1+4n: if line number / 4 leaves a remainder of 1
    if line_number%4==1:
        line=line.rstrip() #strip the newline from each line
        for base in line:#go through each base so we can count it
            counter[base]+=1
            
print(counter)
```

**Count the number of each pair of bases (described as kmers of size 2).**

```python
count_2kmers= {'AA':0,'CC':0,'GG':0,'TT':0,'AT':0,'AC':0, 'AG':0, 'CA':0, 'CG':0,'CT':0, 'GT':0,
               'GA':0,'GC':0,'TA':0,'TC':0, 'TG':0}

for line_number,line in enumerate(fastq):
        if line_number%4==1: #if it's a sequencing read: if it's line 1+4n: if line number/ 4 leaves a remainder of 1
            line=line.rstrip() #strip the newline from each line
            for base_num, base in enumerate(line): #to enumerate the line by each base number
                if base_num != (len(line)-1): #to not include the last base in the line in counting
                    kmers= (line[base_num:base_num+2])# to add the current base to the next base so we can count pairs of bases
                    count_2kmers [kmers]+=1 # add the new kmers (pair of bases) to the coun_2kmers
                
                
print (count_2kmers)
          
            
```

**Count the number of each pair of bases, without  assuming in advance the possible pairs**

```python
count2kmer_unknown={} #start with empty dictionary
for line_number,line in enumerate(fastq):
    if line_number%4==1:
        line=line.rstrip()
        for base_number, base in enumerate(line):
            if base_number != (len(line)-1):
                kmers =(line[base_number:base_number+2])
                if kmers not in count2kmer_unknown: #if new kmers that not in the dictionary added it so we can start cont it
                    count2kmer_unknown [kmers]=1
                else:                             #if the Kmers in the dictionary add 1 for counting
                    count2kmer_unknown [kmers]+=1
                    
                
        
print(count2kmer_unknown)
```

**counting kmers of any size**

```python
count_kmers={} #start with empty dictionary
for line_number,line in enumerate(fastq[0:2]):
    if line_number%4==1:
        line=line.rstrip()
        
        for base_number, base in enumerate(line):
            
            for i in range (len(line)): # because we want to count any size of kmers, i will be any number in range of length of line 
                kmers =(line[base_number:base_number+(i)]) # for kmers size 
                if kmers not in count_kmers:#if new kmers that not in the dictionary added it so we can start cont it
                     count_kmers [kmers]=1
                else:                    #if the Kmers in the dictionary add 1 for counting
                    count_kmers[kmers]+=1
                    
                
        
print(count_kmers)
```
